```python
import re
import json
from bs4 import BeautifulSoup
import requests

URL = "https://www.gumtree.co.za/s-houses-flats-for-sale/johannesburg/v1c9074l3100090p1"
page = requests.get(URL)
print(page.text)
html = page.text
soup = BeautifulSoup(html, 'html.parser')


#  using regex to extract from JSON webpage
pattern = r'var\s+galleryAdList_searchGallery\s*=\s*(\[.*?\]);'
match = re.search(pattern, html, re.DOTALL)

if match:
    json_text = match.group(1)
    listings = json.loads(json_text)

    for listing in listings:
        title = listing.get('title', 'No title')
        price = listing.get('price', {}).get('formattedAmount', 'No price')
        location = listing.get('geo', {}).get('name', 'No location')
        description = BeautifulSoup(listing.get('description', ''), 'html.parser').get_text(separator=' ', strip=True)
        url = "https://www.gumtree.co.za" + listing.get('viewSeoUrl', '')

        print(f"Title: {title}")
        print(f"Price: {price}")
        print(f"Location: {location}")
        print(f"Description: {description}")
        print(f"URL: {url}")
        print()
else:
    print("No listings found ")
    
import pandas as pd
df = pd.DataFrame(listings)
df.to_csv("gumtree_web_scraping_Q1.csv", index=False)
```
